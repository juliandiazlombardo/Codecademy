# Import libraries

```python
import csv
```

# Import the raw data (no pandas)

```python
with open("insurance.csv") as medical_data_file:
    raw_medical_data = csv.DictReader(medical_data_file)
    medical_data = []
    for data in raw_medical_data:
        medical_data.append(data)
```

```python
print(medical_data)
```

```python
def create_dataset(list:list) -> dict:
    keys = list[0].keys()
    dataset = {}
    for key in keys:
        dataset[key] = []
        for data_instance in list:
            dataset[key].append(data_instance[key])
    return dataset
            
medical_dataset = create_dataset(medical_data)
```

```python
for key in medical_dataset:
    print(len(medical_dataset[key]))
```
